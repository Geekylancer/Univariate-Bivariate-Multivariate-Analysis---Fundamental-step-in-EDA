```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('iris.csv')
df.head()
```

**Univariate Analysis**

considering only one feature

```
df[df['species']== 'Iris-setosa'].head(10)
```

```
df[df['species']== 'Iris-versicolor'].head(10)
```

```
df[df['species']== 'Iris-virginica'].head(10)
```

```
df1 = df[df['species']== 'Iris-setosa']
df2 = df[df['species']== 'Iris-versicolor']
df3 = df[df['species']== 'Iris-virginica']
```

```
plt.plot(df1['petal_width'],np.zeros_like(df1['petal_width']),'go')
plt.show()
```

```
plt.plot(df2['petal_width'],np.zeros_like(df2['petal_width']),'bo')
plt.show()
```

```
plt.plot(df3['petal_width'],np.zeros_like(df3['petal_width']),'ro')
plt.show()
```

```
plt.xlabel('Petal Width')
plt.plot(df1['petal_width'],np.zeros_like(df1['petal_width']),'go')
plt.plot(df2['petal_width'],np.zeros_like(df2['petal_width']),'bo')
plt.plot(df3['petal_width'],np.zeros_like(df3['petal_width']),'ro')
plt.show()
```

*CONCLUSION - we find that our output or species column are overlapped with each-other, so we don't go with Univariate analysis*

**Bivariate Analysis**

considering two features

```
sns.set_style('dark')
sns.set(rc={'figure.figsize':(8,5)})
sns.scatterplot(x = 'sepal_length',y ='petal_width',hue = 'species',data = df)
plt.show()
```

```
sns.scatterplot(x = 'petal_length',y ='sepal_width',hue = 'species',data = df)
plt.show()
```

*CONCLUSION - we find that it is easy to classify iris-setosa but classification of the rest species is difficult due to overlapping*

**Multivariate Analysis**

To analysis more than two features (for eg. it is impossible to visualize 4D,5D,... as a human)

```
sns.pairplot(data = df,hue = 'species')
plt.show()
```

*CONCLUSION - when we compare petal_width vs petal_length, we observe a positive correlation and in terms of regression it is a linear one.*
